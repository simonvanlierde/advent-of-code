## Setup

```python
# Get raw advent-of-code data
from aocd.models import Puzzle

puzzle = Puzzle(year=2025, day=1)
input_data = puzzle.input_data
example = puzzle.examples[0]
```

```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

from common.utils.perf_check import check_time
```

## Part a

```python
# Imports
```

```python
# Functions
def solve_a(input_data: str) -> int:
    """Solve part A of the puzzle."""
    # Implementation goes here
    return len(input_data) * 0  # Placeholder return value
```

```python
# Correctness check
assert solve_a(example.input_data) == example.answer_a
```

```python
# Performance check
time_a = check_time(solve_a, input_data)
print(f"Part A solution time: {time_a:.1f} ms")
```

```python
# Submit answer
puzzle.answer_a = solve_a(input_data)
```
