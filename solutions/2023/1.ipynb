## Setup

```python
import sys
from pathlib import Path

from aocd.models import Puzzle

# Add parent directory to path to allow relative imports into Jupyter notebook
sys.path.append(str(Path.cwd().parent))
```

```python
# Get raw advent-of-code data
puzzle = Puzzle(year=2023, day=1)
data = puzzle.input_data
examples = puzzle.examples
```

## Part a

```python
import re
```

```python
solution_part_a = sum(
    int(digits[0] + digits[-1])  # Two-digit number of first and last digit
    for digits in (
        re.findall(r"(\d)", line)  # For each set of digits in a line
        for line in data.splitlines()  # For each line in the data
    )
)
```

```python
# Submit answer
puzzle.answer_a = solution_part_a
```

## Part b

```python
DIGIT_MAP = {
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9",
    "zero": "0",
}
# Following the same pattern as part a:
solution_part_b = sum(
    # Map first and last digit from spelled to number, defaulting to the number itself
    int(DIGIT_MAP.get(digits[0], str(digits[0])) + DIGIT_MAP.get(digits[-1], str(digits[-1])))
    # Expand regex to match spelled out digits
    for digits in (re.findall(r"(?=(\d|" + "|".join(DIGIT_MAP.keys()) + "))", line) for line in data.splitlines())
)
```

```python
# Submit answer
puzzle.answer_b = solution_part_b
```
