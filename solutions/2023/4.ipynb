## Setup

```python
import sys
from pathlib import Path

from aocd.models import Puzzle

# Add parent directory to path to allow relative imports into Jupyter notebook
sys.path.append(str(Path.cwd().parent))
```

```python
# Get raw advent-of-code data
puzzle = Puzzle(year=2023, day=4)
data = puzzle.input_data
examples = puzzle.examples
```

## Part a

```python
solution_part_a = sum(
    # Double the score for each matching number, starting at 1 for the first match.
    2 ** (len({*winning_str.split()} & {*have_str.split()}) - 1) // 1  # Floor division by 1 to ensure integer score
    for line in data.splitlines()  # For each game in the data
    for winning_str, have_str in [line.split(": ")[1].split(" | ")]  # Split the line into the winning and have strings
)
```

```python
# Submit answer
puzzle.answer_a = solution_part_a
```

## Part b

```python
# Initialize card counts
cards = dict.fromkeys(range(1, len(data.splitlines()) + 1), 1)

for i, line in enumerate(data.splitlines(), 1):  # For each card
    w, h = line.split(": ")[1].split(" | ")  # Split the line into the winning and have strings
    matches = len({*w.split()} & {*h.split()})  # Count the number of matching numbers

    # Update copy counts for next 'matches' cards
    for j in range(i + 1, i + matches + 1):
        cards[j] += cards[i]

solution_part_b = sum(cards.values())
```

```python
# Submit answer
puzzle.answer_b = solution_part_b
```
