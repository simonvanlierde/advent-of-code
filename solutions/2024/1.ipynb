## Imports

```python
# Import advent of code lib
from aocd import get_data, submit

# Get raw data
data: str = get_data(year=2024, day=1)
```

## Part a

```python
# Imports
import numpy as np
```

```python
# Load the data into numpy columns
left_col, right_col = np.loadtxt(data.splitlines()).T

# Calculate the sum of the absolute differences between the two sorted columns
total_difference = abs(np.sort(left_col) - np.sort(right_col)).sum()
```

```python
# Submit the result
submit(total_difference, part="a", day=1, year=2024)
```

## Part b

```python
# For each number in the left column, count its frequency in the right column and multiply the two numbers
similarity_score = sum(num * np.count_nonzero(right_col == num) for num in left_col)
```

```python
# Submit the result
submit(similarity_score, part="b", day=1, year=2024)
```
