## Fetch data

```python
# Import advent of code lib
from aocd import get_data, submit

# Get raw data
data: str = get_data(year=2024, day=3)
```

## Part a

```python
# Import regex
import re
```

```python
def get_sum_product_from_text(text: str) -> int:
    """Looks for all the mul(a,b) function calls in a text and returns the sum of the products."""
    return sum(int(a) * int(b) for a, b in re.findall(r"mul\((\d+)\,(\d+)\)", text))
```

```python
# Get the sum of the products from the data
sum_product: int = get_sum_product_from_text(data)
```

```python
# Submit answer
submit(sum_product, part="a", day=3, year=2024)
```

## Part b

```python
enabled_sum_product: int = sum(
    # Get the sum of the products from the text until the first don't() call
    get_sum_product_from_text(do_text.split("don't()")[0])
    # For each do_text in the data
    for do_text in re.split(r"do\(\)", data)
)
```

```python
# Submit answer
submit(enabled_sum_product, part="b", day=3, year=2024)
```
