## Setup

```python
# Get raw advent-of-code data
from aocd.models import Puzzle

puzzle = Puzzle(year=2015, day=1)
input_data = puzzle.input_data
example = puzzle.examples[0]
```

```python
# Import performance checking utility
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

from common.utils.perf_check import check_example, time_solution
```

## Part a

```python
# Functions
def solve_a(input_data: str) -> int:
    """Simply count the number of opening and closing parentheses and return the difference."""
    return input_data.count("(") - input_data.count(")")
```

```python
# Correctness check
check_example(solve_a, example)
```

```python
# Performance check
time_a = time_solution(solve_a, input_data, time_unit="us")
```

```python
# Submit answer
puzzle.answer_a = solve_a(input_data)
```

## Part b

```python
# Functions
def solve_b(input_data: str) -> int:
    """Iterate through the floor changes and return the position where the basement is first reached."""
    f = 0
    for i, c in enumerate(input_data, start=1):
        f += 1 if c == "(" else -1
        if f == -1:
            return i

    # If the basement is never reached, return -1
    return -1
```

```python
# Performance check
time_b = time_solution(solve_b, input_data, time_unit="us")
```

```python
# Submit answer
puzzle.answer_b = solve_b(input_data)
```
