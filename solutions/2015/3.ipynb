## Setup

```python
# Get raw advent-of-code data
from aocd.models import Puzzle

puzzle = Puzzle(year=2015, day=3)
input_data = puzzle.input_data
```

```python
# Import performance checking utility
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

from common.utils.perf_check import time_solution
```

## Part a

```python
# Functions
def solve_a(input_data: str) -> int:
    """Find thenumber of houses that receive at least one present."""
    char_to_move = {"^": 1j, "v": -1j, ">": 1, "<": -1}

    pos = 0
    visited_houses = {pos}

    for move in input_data:
        pos += char_to_move[move]
        visited_houses.add(pos)

    return len(visited_houses)
```

```python
# Performance check
time_a = time_solution(solve_a, input_data)
```

```python
# Submit answer
puzzle.answer_a = solve_a(input_data)
```

## Part b

```python
# Functions
def solve_b(input_data: str) -> int:
    """Find thenumber of houses that receive at least one present, with Santa and Robo-Santa taking turns."""
    char_to_move = {"^": 1j, "v": -1j, ">": 1, "<": -1}

    visited_houses = {0}

    # Split moves between Santa and Robo-Santa
    for moveset in (input_data[::2], input_data[1::2]):
        pos = 0

        for move in moveset:
            pos += char_to_move[move]
            visited_houses.add(pos)

    return len(visited_houses)
```

```python
# Performance check
time_b = time_solution(solve_b, input_data)
```

```python
# Submit answer
puzzle.answer_b = solve_b(input_data)
```
